## 1. Approximating the value of $\pi$ via Monte Carlo method

Let's implement the idea of Monte Carlo method applying to approximate the numerical value of $\pi$. This is a famous and elementary description of the method:<br><br>

consider a unit circle inside a square with four vertices being $(0, 0), (1, 0), (1, 1), (0, 1)$ on a coordinate plane(i.e. $\mathbb{R}^2$). We only need to consider the part on the 1st quadrant. We know that circle part has area $\frac{\pi}{4}$ and square part has area $1$.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
sample_number = 500

x = np.random.rand(sample_number)
y = np.random.rand(sample_number)

in_cir = 0  # count number for randomly generated points inside the unit circle
out_cir = 0

for i in range(sample_number):
    dist = x[i] ** 2 + y[i] ** 2
    if dist <= 1:  # inside circle
        in_cir += 1
    else:
        out_cir += 1

# approximation of pi
4 * in_cir / sample_number

# it is easy to see that the larger sample_number is, the more precise our approximation.
```

```python
# plot to see what's happening

sample_number = 500

x = np.random.rand(sample_number)
y = np.random.rand(sample_number)

dot_in_cirx = []
dot_in_ciry = []

in_cir = 0
out_cir = 0

for i in range(sample_number):
    dist = x[i] ** 2 + y[i] ** 2
    if dist <= 1:
        in_cir += 1
        dot_in_cirx.append(x[i])
        dot_in_ciry.append(y[i])        
    else:
        out_cir += 1

# approximation of pi
4 * in_cir / sample_number

plt.figure()
plt.scatter(x, y)
plt.scatter(dot_in_cirx, dot_in_ciry, color='r')
plt.show()
```
