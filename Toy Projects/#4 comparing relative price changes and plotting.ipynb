# #4. Comparing relative stock price changes and its plot

```python
import yfinance as yf  
import datetime as dt
import matplotlib.pyplot as plt
```

```python
start = dt.date(2020,5,1)
end = dt.date(2020,5,15)
stocks = 'AAPL'
```

```python
df = yf.download(stocks, start, end)
df
```

```python
# preparing data set
start = dt.date(2000, 1, 1)
end = dt.datetime.now()
df = yf.download(stocks, start, end)['Adj Close']
df
```

# Visualization

```python
df.plot()
plt.show()
```

```python
# relative price changes = daily returns
daily_returns = df.pct_change()
daily_returns
```

```python
# plot the volatility
daily_returns.plot()
plt.show()
```

```python
# price comparison with 'MSFT'
stocks = ['AAPL', 'MSFT']
df = yf.download(stocks, start, end)['Adj Close']
df
```

```python
df.plot()
plt.show()
# plot below is misleading: this compares absolute price changes
# instead, need to compare relative price changes
```

```python
daily_returns = df.pct_change()
daily_returns
```

```python
daily_cum_returns = (daily_returns + 1).cumprod()
daily_cum_returns
```

```python
daily_cum_returns.plot()
plt.show()
```

```python
# plot the volatility
daily_returns.plot()
plt.show()
```
